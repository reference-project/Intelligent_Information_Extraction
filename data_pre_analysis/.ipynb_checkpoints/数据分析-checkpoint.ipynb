```python
# 统计训练集中的字符种类个数
max_num = -1
for i in range(len(train_data)):
    temp1 = train_data.iloc[i][0].split('  ')
    for t1 in temp1:
        temp2 = [int(tt) for tt in t1.split('/')[0].split('_')]
        max_num = max(max_num, max(temp2))
print(max_num)
```

```python
min_num = 25000
for i in range(len(train_data)):
    temp1 = train_data.iloc[i][0].split('  ')
    for t1 in temp1:
        temp2 = [int(tt) for tt in t1.split('/')[0].split('_')]
        min_num = min(min_num, min(temp2))
print(min_num)
```

```python
word_freq = [0] * 21225
for i in range(len(train_data)):
    temp1 = train_data.iloc[i][0].split('  ')
    for t1 in temp1:
        temp2 = [int(tt) for tt in t1.split('/')[0].split('_')]
        for t2 in temp2:
            word_freq[t2] += 1 
print(word_freq)
```

```python
word_freq.index(max(word_freq))
```

```python
n = 0
for freq in word_freq:
    if not freq:
        n += 1
print(n)
```

```python
target_freq = [0] * 4
for i in range(len(train_data)):
    temp1 = train_data.iloc[i][0].split('  ')
    for t1 in temp1:
        tt = t1.split('/')[1]
        if tt == 'a':
            target_freq[0] += 1
        elif tt == 'b':
            target_freq[1] += 1
        elif tt == 'c':
            target_freq[2] += 1
        else:
            target_freq[3] += 1
print(target_freq)
```
